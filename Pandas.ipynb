# IMPORTAR NOVA BASE DE DADOS (COM TODOS OS DADOS)

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
```

```python
dados.head()
```

```python
dados.fillna(0, inplace=True)
```

```python
dados['Valor_por_Mes'] = dados['Condominio'] + dados['Valor']
```

```python
dados.head()
```

```python
dados['Valor_por_ano'] = (12 * dados['Valor_por_Mes']) + dados['IPTU']
```

```python
dados.sort_values('Valor_por_ano')
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + dados['Quartos'].astype(str) + ' quarto(s) ' + ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
```

```python
dados
```

```python
dados['Descricao']
```

```python
dados['Possui_Suite'] = dados['Suites'].apply(lambda x: 'Sim' if x > 0 else 'Não')
```

```python
dados.head()
```

```python
dados.to_csv('Dados_completos.csv', index=False, sep=';')
```
