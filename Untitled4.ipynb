```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados_idade_aposentadoria = pd.read_csv('dados_idade_aposentadoria.csv')
dados_vida_lampada = pd.read_csv('dados_vida_lampada.csv')
dados_alturas = pd.read_csv('dados_alturas.csv')
```

```python
plt.figure(figsize=(20, 5))
plt.subplot(131)
plt.title('Distribuicao de Idade de Aposentadoria')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.hist(dados_idade_aposentadoria['idade'], bins=20, color='blue')

plt.subplot(132)
plt.title('Distribuicao de Vida da Lampada')
plt.xlabel('Tempo de Vida (h)')
plt.ylabel('Frequência')
plt.hist(dados_vida_lampada, bins=20, color='magenta')

plt.subplot(133)
plt.title('Distribuicao de Alturas')
plt.xlabel('Altura (cm)')
plt.ylabel('Frequência')
plt.hist(dados_alturas, bins=20, color='purple')

plt.show()
```

```python
# Reamostragens

# tamanho da amostra
n = 100
# quantidade de amostras
qnt = 100000

# função de reamostragem e cálculo de médias
def reamostragem_media(dados, coluna, n, qnt):
    medias = [dados[coluna].sample(n, replace=True).mean() for _ in range(qnt)]
    return medias
```

```python
media_idade = reamostragem_media(dados_idade_aposentadoria, 'idade', n, qnt)
media_duracao = reamostragem_media(dados_vida_lampada, 'duracao', n, qnt)
media_altura = reamostragem_media(dados_alturas, 'alturas', n, qnt)
```

```python
print("*****  Médias populacionais *****")

print("idade:", dados_idade_aposentadoria['idade'].mean())
print("duracao:", dados_vida_lampada['duracao'].mean())
print("alturas:", dados_alturas['alturas'].mean())
```

```python
def Average(lst):
    return sum(lst) / len(lst)

print("*****  Médias pamostrais *****")

print("idade:", Average(media_idade))
print("duração:", Average(media_duracao))
print("altura:", Average(media_altura))
```

```python
plt.subplots(figsize=(15, 5))
plt.subplot(131)
plt.title('Distribuição das Idade médias de Aposentadoria')
plt.xlabel('Idade média de Aposentadoria')
plt.ylabel('Frequência')
plt.hist(media_idade, bins=30, alpha=0.7, color='blue')

plt.subplot(132)
plt.hist(media_duracao, bins=30, alpha=0.7, color='purple')
plt.title('Duração média de uma Lâmpada')
plt.xlabel('Tempo de Vida médio(horas)')

plt.subplot(133)
plt.hist(media_altura, bins=30, alpha=0.7, color='green')
plt.title('Alturas média dos Funcionários')
plt.xlabel('Altura média (cm)')

plt.show()
```

```python
# importando o stats
from scipy import stats
```

```python
# levar em consideração essa amostra
dados_vida_lampada

# nível de confiança definido
confianca = 0.95
```

```python
#informações para o IC

media = dados_vida_lampada['duracao'].mean()
desvio_padrao_amostral = dados_vida_lampada['duracao'].std()
tamanho_amostra = len(dados_vida_lampada)
```

```python
# Calculando intervalo de confiança

intervalo_confianca = stats.norm.interval(confianca,
                                       loc=media ,
                                       scale=desvio_padrao_amostral/np.sqrt(tamanho_amostra))
```

```python
# Visualizando os resultados

print("IC (95%):", intervalo_confianca)
```
